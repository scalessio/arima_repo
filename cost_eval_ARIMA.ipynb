```python
    
res2 = arima_series.join(ground_trut)
res2.columns=['ARIMA','Ground_Truth']
res2 = res2.tail(18)
res2
```

```python
delta = DataFrame()
delta['Delta'] = res2['ARIMA']-res2['Ground_Truth']
res2 = res2.join(delta)

df=DataFrame()
df['Bin_avaiability_gt'] = res2['Ground_Truth']%9000
res2 = res2.join(df)
res2
```

```python
def func(bin_avaia, delta,lstm_forecast):
    capacity_nfv=9000
    if delta >= 0:
            if bin_avaia >= delta : return ceil(delta/capacity_nfv) 
            else : return 0
    else : 
        avaiability = ((ceil(lstm_forecast%capacity_nfv)*capacity_nfv)-lstm_forecast) % 9000
        if avaiability >= abs(delta) : return 0
        else : return -(ceil(abs(delta)/capacity_nfv))
```

```python
error = DataFrame()
error['NFV_error'] = [ func(res2.iloc[x,3],res2.iloc[x,2],res2.iloc[x,0]) for x in range(0,len(res2))]
error.head(5)
```
