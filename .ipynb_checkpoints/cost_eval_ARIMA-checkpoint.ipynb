```python
import pandas as pd
from pandas import read_csv
from pandas import DatetimeIndex
from pandas import datetime
from pandas import Series
from pandas import concat
from pandas import groupby
from pandas import DataFrame
from math import sqrt
from matplotlib import pyplot
import numpy as np
from numpy import concatenate
from natsort import natsorted, ns
import seaborn as sns
import csv
import glob, os
from math import ceil
```

```python
def Error_Eval(bin_avaia, delta,lstm_forecast):
    capacity_nfv=9000
    if delta >= 0:
            if bin_avaia >= delta : return ceil(delta/capacity_nfv) 
            else : return 0
    else : 
        avaiability = ((ceil(lstm_forecast%capacity_nfv)*capacity_nfv)-lstm_forecast) % 9000
        if avaiability >= abs(delta) : return 0
        else : return -(ceil(abs(delta)/capacity_nfv))
        
```

# Begin


```python
series = DataFrame()
series = read_csv('results/arima_prediction600_lg5.csv', header=0, index_col='date_time', sep='\t',squeeze=True)
ground_trut =read_csv('results/test_prediction600_lg5.csv', header=0, index_col='date_time', sep='\t',squeeze=True)
idxa = series.index
```

```python
arima_series = DataFrame(columns=['ARIMA'],index=idxa)
s = []
ser = series.values
for x in range(0,len(series)):
    s.append(ser[x])
arima_series['ARIMA']=s
res2 = arima_series.join(ground_trut)
res2.columns=['ARIMA','Ground_Truth']
```

```python
#Important: This paramiter depends on number of samples of forecast
res2 = res2.tail(19998)

ind = []

for x in range(0,len(idxa),3):
    ind.append(idxa[x])
    
len(ind)
summarize_ground=[]
summarize = []
for x in range(0,len(res2),3):
    summarize.append(res2.iloc[x:x+3,0].sum())
    summarize_ground.append(res2.iloc[x:x+3,1].sum())
    
len(summarize)
```

```python
res2 = DataFrame(index=ind)    
    
res2['ARIMA'] = summarize
res2['Ground_Truth'] = summarize_ground
res2
```

```python
#res2 = res2.tail(19998)
delta = DataFrame()
df=DataFrame()

delta['Delta'] = res2['ARIMA']-res2['Ground_Truth']
res2 = res2.join(delta)

df['Bin_avaiability_gt'] = res2['Ground_Truth']%9000
res2 = res2.join(df)
res2
```

```python

error = DataFrame()
error['NFV_error'] = [ Error_Eval(res2.iloc[x,3],res2.iloc[x,2],res2.iloc[x,0]) for x in range(0,len(res2))]

pyplot.plot(range(len(error.NFV_error)), error.NFV_error, color='y', alpha=1.0,  ms=3, mfc='b', mec='b', label='Forecast')
pyplot.rcParams['figure.figsize'] = (12,9)
#pyplot.title('NF Error Provisioning using ARIMA', weight='bold',fontsize=20)
pyplot.ylabel("NF Overprovisioning / Undeprovisioning ",fontsize=24)
pyplot.xlabel("Forecast_minutes",fontsize=24)
pyplot.yticks()
#axes.set_xlim([-1,7000])
#axes.set_ylim([-20,+5])
pyplot.savefig('plots/cost_evaluation/NF_err_provisioning.png')
pyplot.show()
```

```python
pyplot.plot(range(len(error.NFV_error[-20:])), error.NFV_error[-20:], color='y', alpha=1.0,  ms=3, mfc='b', mec='b', label='Forecast')
pyplot.rcParams['figure.figsize'] = (12,9)
#pyplot.title('Zoom NF Error Provisioning Using Arima', weight='bold',fontsize=20)
pyplot.ylabel("NF Overprovisioning / Undeprovisioning",fontsize=24)
pyplot.xlabel("Forecast_minutes",fontsize=24)
pyplot.yticks()
#axes = pyplot.gca()
#axes.set_xlim([-1,7000])
#axes.set_ylim([-20,+5])
pyplot.savefig('plots/cost_evaluation/Zoom_NF_err_provisioning.png')
pyplot.show()
```

```python
a = error.NFV_error.values
unique, counts = np.unique(a, return_counts=True)
y_pos = np.arange(len(unique))
performance =counts
 
pyplot.bar(y_pos, performance, align='center',color=['red', 'green', 'red'], alpha=0.5, label='ARIMA')
pyplot.legend(loc='upper left', fancybox=True, fontsize='large', framealpha=0.5)
pyplot.xticks(y_pos, unique)
pyplot.xlabel('Underprovisioning/Overprovisioning',fontsize = 24)
pyplot.ylabel('Frequency',fontsize = 24)
#pyplot.title('Error Of last 6000 forecasts Using ARIMA', weight='bold',fontsize = 20)
pyplot.savefig('plots/cost_evaluation/bar_ERROR_provisioning.png')
pyplot.show()
```
